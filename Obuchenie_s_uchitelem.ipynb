# Отток клиентов

## Введение

Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых. «Бета-Банк» предоставил исторические данные о поведении клиентов и расторжении договоров с банком. Маркетологи хотят спрогнозировать, уйдёт клиент из банка в ближайшее время или нет.

**Задача.**
Необходимо построить модель с предельно большим значением *F1*-меры (не менее 0.59). Проверить *F1*-меру на тестовой выборке.

**Подготовка данных.**
Для решения задачи считаем и изучим предоставленные данные: проверим названия столбцов, наличие пропусков, аномальных значений; проверим данные на дубликаты. Также необходимо проверить численные данные на выбросы. Для этого к каждому численному столбцу построим гистограмму и боксплот для анализа.


**Исследование задачи.**
После того, как данные будут подготовлены к работе, приступим к исследованию задачи. Датафрейм поделим на две выборки (обучающую и тестовую) в соотношении 4:1, преобразуем категориальные значения по одной из техник кодирования, стандартизируем необходимые значения, изучим баланс классов в целевом признаке.

**Обучение модели без учёта дисбаланса.** Обучим модели без учета дисбаланса. Будут обучены следующие модели: решающее дерево, случайный лес, логистическая регрессия. Перед обучением подберём оптимальное значение гиперпараметров для каждой модели, с учётом, что нам нужно получить наилучшую оценку метрики F1. Оптимальное значение подберём методом перекрестной проверки. Для каждой модели построим графики со средними значениями F1 и Accuracy при переборе разных сочетаний гиперпараметров, чтобы посмотреть разницу между значениями F1 и Accuracy в несбаллансированном наборе данных.

**Борьба с дисбалансом.** К каждой модели применим три способа дисбаланса. Понаблюдаем на графиках, как меняется разница между значениями F1 и Accuracy. Найдём модель с наилучшим значением F1.

**Тестирование модели.** Проанализировав, какая модель и с каким способом устранения дисбаланса выдаёт большую долю верных ответов, протестируем её на тестовой модели. Если доля верных ответов будет не меньше 0.59, значит, модель можно использовать на практике.

Создадим функции для построения графиков, которые помогут определить выбросы в данных.

```python
df = data.copy()
```

```python
def boxplot_graf(data, name):
    box = plt.boxplot(
        x=data, vert=False, 
        patch_artist=True,  
        boxprops = dict(facecolor = "#35c0cd")
    )
    fig = plt.gcf()
    fig.set_size_inches(20, 5)
    plt.title(
            f'Боксплот признака {name}', 
            fontsize=14
        )
    plt.show()
    print()
    z1 = data.quantile(0.25)
    z3 = data.quantile(0.75)
    r = z3 - z1
    lower = z1 - (1.5 * r)
    upper = z3 + (1.5 * r)
    print()
    print('\033[1m' + f'Пороговые значения {name}:' + '\033[0m')
    print('Нижняя граница значений:', lower)
    print('Верхняя граница значений:', upper)
    print()
    print()
```

```python
def hist_graf(data, name, title, y_name, x_name, bins):
    print()
    print('\033[1m' + f'Описание признака {name}:' + '\033[0m')
    print()
    print(data.describe())
    print()
    
    data.plot(
        kind='hist', 
        title=title, 
        bins=bins, 
        grid=True, 
        figsize = (15,8),
        color = '#35c0cd',
        edgecolor='black');
    plt.ylabel(y_name)
    plt.xlabel(x_name);
```

### Обучение модели без учёта дисбаланса

Обучим модель без учёта дисбаланса. Создадим функцию перебора гиперпараметров. Accuracy не очень хорошо работает с несбалансированными наборами данных, что часто приводит к вводящим в заблуждение результатам. Посмотрим разницу между F1 и Accuracy на графиках. 

```python
scorers = {
    'precision_score': make_scorer(precision_score),
    'recall_score': make_scorer(recall_score),
    'accuracy_score': make_scorer(accuracy_score),
    'f1_score': make_scorer(f1_score)
}

'''
Функция для перебора гиперпараметров
params_set - наборы гиперпараметров
scorers - метрики классификации, задаются один раз и не меняются.
target - выборка с целевым признаком
features - выборка с остальными признаками
model - вид модели
'''

def G_S_CV (params_set, scorers, cv, features, target, model):
    Grid_CV_DTC = GridSearchCV (
        model, 
        refit='f1_score',
        param_grid = params_set, 
        scoring = scorers, 
        cv = cv, 
        n_jobs = -1
    )
    pipe_grid = Grid_CV_DTC.fit(features, target)
    
    r = pd.DataFrame(pipe_grid.cv_results_) 
#    plt.hist(r["mean_test_f1_score"], bins = 10, alpha = 0.5, label='a', figsize = (15,8))
#    plt.hist(r["mean_test_accuracy_score"], bins = 10, alpha = 0.5, label='b', figsize = (15,8))
#    plt.legend(loc='upper left')

    plt.show()
    r["mean_test_f1_score"].plot(kind='hist', figsize = (15,8), color = '#35c0cd', edgecolor='black', label='Средние значения F1')
    r["mean_test_accuracy_score"].plot(kind='hist', figsize = (15,8), color = '#e1c2be', edgecolor='black', label='Средние значения Accuracy')
    plt.legend(loc='upper left')
    plt.xlabel("Средние значения")
    _ = plt.title("Разница между средними значениями F1 и Accuracy")
    
    best_hyp_DTC = pipe_grid.best_params_
    best_score_DTC = pipe_grid.best_score_
    print(f"Лучшие значения гиперпараметров: {best_hyp_DTC}")
    print(f"f1 наилучшей модели:{best_score_DTC}")
```

#### Решающее дерево

```python
params_set_DTC = {
    'max_depth': [8,15,22],
    'min_samples_leaf': [4,8,15,22],
    'min_samples_split': [4,8,15,22],
}

model_DTC = DecisionTreeClassifier(random_state=12345) 
G_S_CV (params_set_DTC, scorers, 3, features_train, target_train, model_DTC)
```

### Увеличение выборки

```python
# увеличим выборку в 4 раза, так как отрицательный класс больше положительного в 4 раза
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    # сцепляем выборки
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    # перемешиваем значения, чтобы они не шли друг за другом 
    features_upsampled = shuffle(features_upsampled, random_state=12345)
    target_upsampled = shuffle(target_upsampled, random_state=12345)
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 4)

print(features_upsampled.shape)
print(target_upsampled.shape)
```

### Уменьшение выборки

```python
# уменьшим выборку в 4 раза, так как положительный класс больше отрицательного в 4 раза
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    # fractioning samples
    features_downsampled = pd.concat([features_zeros.sample(frac=fraction, random_state=12345)] +
                                     [features_ones])
    target_downsampled = pd.concat([target_zeros.sample(frac=fraction, random_state=12345)] +
                                   [target_ones])
    
    features_downsampled = shuffle(features_downsampled, random_state=12345)
    target_downsampled = shuffle(target_downsampled, random_state=12345)

    return features_downsampled, target_downsampled

features_downsampled, target_downsampled = downsample(features_train, target_train, 0.25)

print(features_downsampled.shape)
print(target_downsampled.shape)
```

## Тестирование модели

Протестируем модель методом Случайный лес с увеличением выборки. Попробуем улучшить результат, изменив пороговое значение. Визуализируем кривую precision_recall и кривую ROC.

```python
'''
Функция перебора пороговых значений.
probabilities_valid - вероятности классов
probabilities_one_valid - значения вероятностей класса «1» 
target - выборка с целевым признаком
'''
def porog (probabilities_valid, probabilities_one_valid, target):
    for threshold in np.arange(0, 0.8, 0.02):
        predicted_train = probabilities_one_valid > threshold
        precision = precision_score(target, predicted_train)
        recall = recall_score(target, predicted_train)
        f1 = f1_score(target, predicted_train)

        print("Порог = {:.2f} | Точность = {:.3f}, Полнота = {:.3f}, f1 = {:.3f}".format(
            threshold, precision, recall, f1))    
```

```python
'''
Функция построения кривой Precision-Recall
target - выборка с целевым признаком
'''

def PR (target):
    precision, recall, thresholds = precision_recall_curve(target, probabilities_valid[:, 1])

    plt.figure(figsize=(8, 8))
    plt.step(recall, precision, where='post', color = '#35c0cd')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.05])
    plt.title('Кривая Precision-Recall')
    plt.show() 
```

```python
'''
Функция построения ROC-кривой 
target - выборка с целевым признаком
'''

def roc (target):
    fpr, tpr, thresholds = roc_curve(target, probabilities_one_valid)
    plt.figure(figsize=(8, 8))
    plt.plot(fpr, tpr, color = '#35c0cd')

    # ROC-кривая случайной модели
    plt.plot([0, 1], [0, 1], linestyle='--')

    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])

    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')

    plt.title('ROC-кривая')

    plt.show()
    
    auc_roc = roc_auc_score(target, probabilities_one_valid)
    print()
    print ('AUC_ROC = ', auc_roc) 
```
